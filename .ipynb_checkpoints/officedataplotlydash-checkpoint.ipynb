```python
import dash
import dash_core_components as dcc
import dash_html_components as html 
import plotly.express as px
from dash.dependencies import Input, Output, State



app = dash.Dash()

app.layout = html.Div([

    html.H1('Here is my Demo Dash board'),

    html.Div([
       
    
     dcc.Dropdown(id = 'dd1',
                 #replacement = 'Select reference data',
                 options = [{'label': i ,'value': i } for i in df_dates.unique()]
                 
                 )
      
      ]),
    
    dcc.Graph(id = 'mg1')
    
       
    
])

@app.callback(
    
Output(component_id = 'mg1', component_property = 'figure' ),
Input (component_id = 'dd1', component_property = 'value'  )

)




def update_graph(dates):
       
    filtered_data = df [df_dates == dates],
                
    fig =  px.scatter(x = df['delta_marginal_cost']['value'], y = df ['delta_fuel']['value']),
                      
    fig.update_layout(transition_duration = 500),
                        
                        
    return fig

                        
                        
                        

                        
if __name__=='__main__':
                          
      app.run_server(port = 4040)                  
                    
                        
                        
                        
                        
                    
                        
                        
                        
    
    



```
